```python
import csv
from xml.etree import ElementTree

# Open Burp XML file and parse it
with open('http_history.xml', 'r') as burp_file:
       burp_data = ElementTree.parse(burp_file).getroot()

# Create CSV file and write header row
with open('http_history.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Timestamp', 'URL', 'Method', 'Status', 'Length', 'MIME Type', 'Request Headers', 'Request Body', 'Response Headers', 'Response Body'])

    # Iterate through HTTP messages and write data rows
    for item in burp_data.iter('item'):
        # Extract data from HTTP message
        timestamp = item.find('time')
        if timestamp is not None:
            timestamp = timestamp.text
        else:
            timestamp = ''
        
        url = item.find('url')
        if url is not None:
            url = url.text
        else:
            url = ''
        
        method = item.find('method')
        if method is not None:
            method = method.text
        else:
            method = ''
        
        status = item.find('status')
        if status is not None:
            status = status.text
        else:
            status = ''
        
        length = item.find('responselength')
        if length is not None:
            length = length.text
        else:
            length = ''
        
        mime_type = item.find('mimetype')
        if mime_type is not None:
            mime_type = mime_type.text
        else:
            mime_type = ''
        
        request_headers = item.find('request/headers')
        if request_headers is not None:
            request_headers = request_headers.text
        else:
            request_headers = ''
        
        request_body = item.find('request/body')
        if request_body is not None:
            request_body = request_body.text
        else:
            request_body = ''
        
        response_headers = item.find('response/headers')
        if response_headers is not None:
            response_headers = response_headers.text
        else:
            response_headers = ''
        
        response_body = item.find('response/body')
        if response_body is not None:
            response_body = response_body.text
        else:
            response_body = ''

        # Write data row to CSV file
        writer.writerow([timestamp, url, method, status, length, mime_type, request_headers, request_body, response_headers, response_body])
```
